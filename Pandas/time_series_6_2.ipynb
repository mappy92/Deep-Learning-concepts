```python
import pandas as pd
df = pd.read_csv("fb.csv",parse_dates=['Date'],index_col='Date')
df
```

```python
df.shift(1) # shifting the price column by one row downwards
```

```python
df.shift(-1) # upwards
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['1 day Price Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['5 day return'] =  (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

```python
df = df[['Price']]
df
```
